<h2>k-NN</h2>

Distancia de Euclides <br>
Geometria do Táxi <br>
Distancia Minkowiski <br>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.iloc[:, 0:4].values #features
y = df[["target"]].values.ravel() #classes
```

```python
knn = KNeighborsClassifier(n_neighbors=13, p=1) ##p=2 -> Euclidana
```

```python
knn.fit(X, y)
```

```python
flor1 = [5.3, 4.7, 2.1, 2.0]
flor2 = [1, 1, 1, 1]

flores = [flor1, flor2]
```

```python
knn.predict(flores) #Nova IRIS
```
